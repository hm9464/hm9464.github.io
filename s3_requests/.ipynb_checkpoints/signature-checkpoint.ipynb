### For policy hash

```python
import base64

input = open("policy.txt", "rb")
policy = input.read()
policy_encoded = base64.b64encode(policy)

print("%s" % (policy_encoded))
```

### For signature

```python
import base64
import hmac
from hashlib import sha1
UTF8 = "utf-8"
my_secret_key = "EN2PRkIVA6c81YY8HAKcXECAQvNR6T0NJnxb0v7m"

def create_signature(secret_key, encoded_policy):
    new_hmac = hmac.new(bytes(secret_key, UTF8), digestmod=sha1)
    new_hmac.update(bytes(encoded_policy, UTF8))
    signature_base64 = base64.b64encode(new_hmac.digest())
    signature = str(signature_base64, UTF8).strip()
    return signature

if __name__ == "__main__":
    my_encoded_policy = "eyJleHBpcmF0aW9uIjogIjIwMjEtMDEtMDFUMDA6MDA6MDBaIiwKImNvbmRpdGlvbnMiOiBbCnsiYnVja2V0IjogImhpbWktZGVlcC1sZWFybmluZyJ9LApbInN0YXJ0cy13aXRoIiwgIiRrZXkiLCAiIl0sCnsiYWNsIjogInByaXZhdGUifSwKeyJzdWNjZXNzX2FjdGlvbl9yZWRpcmVjdCI6ICJpbWFnZV9jbGFzc2lmaWVyLmh0bWwifSwKWyJzdGFydHMtd2l0aCIsICIkQ29udGVudC1UeXBlIiwgIiJdLApbImNvbnRlbnQtbGVuZ3RoLXJhbmdlIiwgMCwgMTA0ODU3Nl0KXQp9"
    signature = create_signature(my_secret_key, my_encoded_policy)
    print(signature)
```
